```python
import pandas as pd
import numpy as np
```

```python
col_list = ['reviews.username','id','name','reviews.rating']
ratings = pd.read_csv("./amazon.csv", usecols=col_list)
ratings
```

```python
prdRatings = ratings.pivot_table(index=['reviews.username'],columns=['id'],values='reviews.rating',fill_value=0)
prdRatings.head()
```

```python
prdStats = ratings.groupby('id').agg({'reviews.rating': [np.size, np.mean]})
prdStats
popularPrdTmp = prdStats['reviews.rating']['size'] >= 5
popularPrd = prdStats[popularPrdTmp].sort_values([('reviews.rating', 'mean')], ascending=False)
popularPrd
```

```python
def recommendItem(sItemNm, dfPivotRating,dfPopularItems):
    
    inferenceRatings = dfPivotRating[sItemNm] 
    inferenceRatings.head()

    similarItems = dfPivotRating.corrwith(inferenceRatings,axis=0)
    similarItems = similarItems.dropna()
    similarItems = similarItems.sort_values(ascending=False)
    
    df = dfPopularItems.join(pd.DataFrame(similarItems, columns=['similarity']))
    df = df.sort_values(['similarity'], ascending=False)
    return df
```

```python
item = 'AVphgVaX1cnluZ0-DR74'
dfTmp = recommendItem(item,prdRatings,popularPrd)

dfTmp['title'] = 'none'

for idx, row in dfTmp.iterrows():  
    tmp01 = ratings.loc[ratings['id'] == row.name].iloc[0]
    dfTmp.loc[idx,'title']= tmp01['name']

dfTmp
```
