```python
import pandas as pd
import numpy as np
```

```python
r_cols = ['user_id', 'prd_id','title', 'rating']
ratings = pd.read_csv('GAME1000.csv', sep=',', names=r_cols, usecols=range(4), encoding="ISO-8859-1",skiprows=1)
ratings
```

```python
#ratings =ratings.drop_duplicates()

#ratings
```

```python
prdRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating',fill_value=0)

prdRatings.head()

```

```python
title = "FF7" 
inferenceRatings = prdRatings[title] 
inferenceRatings.head()

similarPrd = prdRatings.corrwith(inferenceRatings,axis=0)
similarPrd = similarPrd.dropna()
df = pd.DataFrame(similarPrd)
df.head(10)
```

```python
similarPrd.sort_values(ascending=False)
```

```python
prdStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
prdStats.head()
```

```python
popularPrd = prdStats['rating']['size'] >= 1
prdStats[popularPrd].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
df = prdStats[popularPrd].join(pd.DataFrame(similarPrd, columns=['similarity']))
```

```python
df.head()
```

```python
df.sort_values(['similarity'], ascending=False)[:15]
```

```python
inferenceRatings01=prdRatings.loc['Adaline'] 

similarUser = prdRatings.corrwith(inferenceRatings01,axis=1)
similarUser = similarUser.dropna()
similarUser.sort_values(ascending=True)
df = pd.DataFrame(similarUser)
df.head(10)
```
