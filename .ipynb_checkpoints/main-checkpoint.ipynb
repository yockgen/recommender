```python
import pandas as pd
import numpy as np
```

```python
r_cols = ['user_id', 'prd_id','title', 'rating']
ratings = pd.read_csv('GAME1000.csv', sep=',', names=r_cols, usecols=range(4), encoding="ISO-8859-1",skiprows=1)
#ratings
```

```python
ratings =ratings.drop_duplicates()
#atings
```

```python
prdRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating',fill_value=0)
prdRatings.head()

```

```python
def recommendItem(sItemNm, lstPivotRating,lstRawRating,iReccurenceCount):
    
    inferenceRatings = lstPivotRating[sItemNm] 
    inferenceRatings.head()

    similarItems = lstPivotRating.corrwith(inferenceRatings,axis=0)
    similarItems = similarItems.dropna()
    similarItems = similarItems.sort_values(ascending=False)
    
    itemsStats = lstRawRating.groupby('title').agg({'rating': [np.size, np.mean]})
    
    popularItems = itemsStats['rating']['size'] >= iReccurenceCount
    itemsStats[popularItems].sort_values([('rating', 'mean')], ascending=False)
    
    df = itemsStats[popularItems].join(pd.DataFrame(similarItems, columns=['similarity']))
    df = df.sort_values(['similarity'], ascending=False)
    return df
   
    
    
dfTmp = recommendItem("RDR2",prdRatings,ratings,1)
print(dfTmp.head())
```
