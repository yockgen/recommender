```python
import pandas as pd
```

```python
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
```

```python
ratings.head()
```

```python
movieRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movieRatings.head()
```

```python

title = "Forrest Gump (1994)" #Toy Story (1995),Silence of the Lambs, The (1991) #Star Wars (1977)
inferenceRatings = movieRatings[title] 
inferenceRatings.head()
```

```python
similarMovies = movieRatings.corrwith(inferenceRatings)
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False)
```

```python
import numpy as np
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

```python
popularMovies = movieStats['rating']['size'] >= 150
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))
```

```python
df.head()
```

```python
df.sort_values(['similarity'], ascending=False)[:15]
```
